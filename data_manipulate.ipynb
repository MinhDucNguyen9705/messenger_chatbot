```python
def score_round(x):
    if '.' not in str(x):
        return x
    x = str(x)
    int_part = x[:x.index('.')]
    float_part = x[x.index('.')+1:]
    if len(float_part)==1:
        float_part+='0'
    round_list = ['0', '25', '50', '75', '100']
    closest = 0
    minimum = int(float_part)-int(round_list[0])
    for i in  range (len(round_list)):
        if abs(int(float_part)-int(round_list[i]))<minimum:
            minimum =  abs(int(float_part)-int(round_list[i]))
            closest=i
    float_part = round_list[closest]
    if float_part=='100':
        int_part=str(int(int_part)+1)
        float_part = '0'
    return float(int_part+'.'+float_part)

score_data['ngu_van'] = score_data['ngu_van'].apply(score_round)
```

```python
def aggregate_score(agg_col, col_1, col_2, col_3):
    global score_data
    temp_score = score_data.loc[(~score_data[col_1].isnull()) & (~score_data[col_2].isnull()) & (~score_data[col_3].isnull()),[col_1, col_2, col_3]]
    temp_score[agg_col] = temp_score[col_1]+temp_score[col_2]+temp_score[col_3]
    score_data = score_data.merge(temp_score, on=['sbd', col_1, col_2, col_3], how='left')
```

```python
aggregate_score('B00_agg', 'toan', 'hoa_hoc', 'sinh_hoc')
```

```python
aggregate_score('C00_agg', 'ngu_van', 'lich_su', 'dia_li')
```

```python
aggregate_score('A00_agg', 'toan', 'vat_li', 'hoa_hoc')
```

```python
def aggregate_score_fl(agg_col, col_1, col_2, col_3, fl):
    global score_data
    temp_score = score_data.loc[(~score_data[col_1].isnull()) & (~score_data[col_2].isnull()) & (~score_data[col_3].isnull()) & (score_data['ma_ngoai_ngu']==fl),[col_1, col_2, col_3]]
    temp_score[agg_col] = temp_score[col_1]+temp_score[col_2]+temp_score[col_3]
    score_data = score_data.merge(temp_score, on=['sbd', col_1, col_2, col_3], how='left')
```

```python
aggregate_score_fl('D00_agg', 'toan', 'ngu_van', 'ngoai_ngu', 'N1')
```

```python
aggregate_score_fl('A01_agg', 'toan', 'vat_li', 'ngoai_ngu', 'N1')
```
